Objetivo: carregar os dados de vendas (que servirão de base "Actual") e

## 1. Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

%matplotlib inline
sns.set_style("whitegrid")
print("Imports OK")
```

## 2. Carregar o dataset

```python
PROJECT_ROOT = Path.cwd().parent
RAW_PATH     = PROJECT_ROOT / "data" / "raw" / "Sample_Superstore.csv"

df = pd.read_csv(RAW_PATH, parse_dates=["Order Date"], encoding="latin1")
print(f"Shape: {df.shape[0]:,} linhas × {df.shape[1]} colunas")
df.head()
```

## 3. Entendendo as colunas


| Coluna | Significado |
|---|---|
| Order Date | Data da venda |
| Region | Região do pedido |
| Category | Categoria do produto (Furniture, Office Supplies, Technology) |
| Sub-Category | Subcategoria |
| Sales | Valor da venda (R$) — este será nosso "Actual" |
| Profit | Lucro da venda |

Neste projeto, tratamos `Sales` agregado por mês e categoria como o
**valor Realizado (Actual)** — o que de fato aconteceu. Depois vamos
construir um **Orçamento (Budget)** e comparar os dois.


## 4. Diagnóstico de qualidade

```python
print("Tipos de dados:")
print(df.dtypes.to_string())
print()
print("Nulos:", df.isnull().sum().sum())
print()
print(f"Período: {df['Order Date'].min().date()} a {df['Order Date'].max().date()}")
print(f"Categorias: {df['Category'].unique().tolist()}")
```

## 5. Visão geral de vendas por categoria

```python
cat_summary = df.groupby("Category").agg(
    total_sales=("Sales", "sum"),
    total_profit=("Profit", "sum"),
    n_orders=("Row ID", "count")
).round(2)
print(cat_summary.to_string())

fig, ax = plt.subplots(figsize=(7,4))
cat_summary["total_sales"].plot(kind="bar", ax=ax, color="#4C72B0", edgecolor="white")
ax.set_title("Vendas totais por categoria", fontsize=12, pad=10)
ax.set_ylabel("Vendas (R$)")
ax.tick_params(axis="x", rotation=0)
plt.tight_layout()
(PROJECT_ROOT / "docs" / "screenshots").mkdir(parents=True, exist_ok=True)
plt.savefig(PROJECT_ROOT / "docs" / "screenshots" / "01_sales_by_category.png", dpi=150)
plt.show()
```
